```python
import re

def clean_text_only(df, text_field, new_text_field_name):
    df[new_text_field_name] = df[text_field].apply(
        lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+://\S+)|^rt|http.+?", "", elem)
    )
    # Menghapus angka
    df[new_text_field_name] = df[new_text_field_name].apply(
        lambda elem: re.sub(r"\d+", "", elem)
    )
    return df
```

```python
my_df = clean_text_only(my_df, 'content', 'text_cleaned')
my_df['text_cleaned']
```

```python
def case_fold_text(df, text_field, new_text_field_name):
    df[new_text_field_name] = df[text_field].str.lower()
    return df
```

```python
#import library re, praproses
import re
def praproses(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)()(?:-)?(?:\)|\(|D|P)',
                           text)
    text = (re.sub('[\W]+', ' ', text.lower()) +
            ' '.join(emoticons).replace('-', ''))
    return text
```

```python
from imblearn.over_sampling import RandomOverSampler

vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(my_df['content'])

ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X_vectorized, my_df['Label'])

X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.20, random_state=0, stratify=y_resampled
)

clf = MultinomialNB()
clf.fit(X_train, y_train)
predicted = clf.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print("MultinomialNB Accuracy:", accuracy_score(y_test,predicted))
print(f"confusion_matrix:\n{confusion_matrix(y_test, predicted)}")
print(classification_report(y_test, predicted, zero_division=0))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Buat classification report dalam bentuk dictionary
report = classification_report(y_test, predicted, output_dict=True, zero_division=0)

# Hitung akurasi total
accuracy = accuracy_score(y_test, predicted)

# Kelas dan metrik yang akan ditampilkan
labels = ['Negatif', 'Netral', 'Positif']
metrics = ['precision', 'recall', 'f1-score']

# Buat DataFrame dari report
data = {
    'Metric': [],
    'Value': [],
    'Label': []
}

for label in labels:
    for metric in metrics:
        data['Metric'].append(metric.capitalize())
        data['Value'].append(report[label][metric])
        data['Label'].append(label)

# Tambahkan akurasi ke DataFrame
data['Metric'].append('Accuracy')
data['Value'].append(accuracy)
data['Label'].append('Total')

metrics_df = pd.DataFrame(data)

# Buat grafik batang
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='Metric', y='Value', hue='Label', data=metrics_df, palette='Set2')

# Judul dan label sumbu
plt.title('Evaluasi Model Negatif, Netral, Positif dan Akurasi Total', fontsize=14)
plt.ylim(0, 1)
plt.ylabel('Nilai Metrik')
plt.xlabel('Metrik Evaluasi')

# Tambahkan nilai angka di atas batang
for p in barplot.patches:
    height = p.get_height()
    if height > 0:
        barplot.annotate(format(height, '.2f'),
                         (p.get_x() + p.get_width() / 2., height),
                         ha='center', va='bottom',
                         fontsize=8, color='black',
                         xytext=(0, 5),
                         textcoords='offset points')

# Tampilkan legenda dengan judul
plt.legend(title='Kelas')

# Tata letak rapi
plt.tight_layout()
plt.show()

# Cetak classification report lengkap dan akurasi
print(f"\nAkurasi Model: {accuracy:.4f}")
print("\nClassification Report Lengkap:")
print(classification_report(y_test, predicted, zero_division=0))
```
